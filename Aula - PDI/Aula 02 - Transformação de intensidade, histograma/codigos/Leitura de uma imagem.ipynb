# Leitura e plotagem de imagens utilizando Python

```python
import matplotlib.pyplot as plt   # Biblioteca matplotlib, utilizada para plotar figuras. Pode
                                  # ser usada tamb√©m para ler imagens do disco
import numpy as np                # Biblioteca numpy
```

```python
# Leitura da imagem utilizando o matplotlib. A imagem retornada √© um array da biblioteca numpy
img = plt.imread('earth.jpg')
# Plotagem da imagem
plt.imshow(img)
```

√â sempre interessante verificarmos algumas informa√ß√µes sobre a imagem lida:

```python
print('Objeto:', type(img))
print('Tipo de dado de cada pixel:', img.dtype)
print('Tamanho da imagem:', img.shape)
print('Valor m√°ximo:', img.max())
print('Valor m√≠nimo:', img.min())
```

______________________________
No caso, a imagem √©:
* Do tipo unsigned integer de 8 bits
* Possui 800 linhas e 1200 colunas
* Possui 3 canais

```python
# Podemos imprimir o valor de um pixel espec√≠fico da imagem:
print(img[400, 560])
```

# Extra√ß√£o e visualiza√ß√£o de um canal de cor

```python
img_r = img[:,:,0]   # Sintaxe do numpy para extra√ß√£o do canal 0 (vermelho)
plt.imshow(img_r)
```

Porque obtemos essa imagem azul e verde?

Ao visualizarmos uma matriz de valores no matplotlib, a biblioteca n√£o sabe que queremos visualizar uma imagem. Por padr√£o, a biblioteca mapeia os valores da matriz usando um **mapa de cores**. Esse mapa de cores gera cores falsas para representar os valores da matriz. Para o que queremos fazer, √© mais interessante visualizar os valores em n√≠veis de cinza. Para isso, fazemos o seguinte:

```python
plt.imshow(img_r, cmap='gray')
```

Vemos que regi√µes da imagem original nas quais o vermelho √© mais predominante (continentes) se tornam brancas nessa imagem em n√≠vel de cinza do canal vermelho.
